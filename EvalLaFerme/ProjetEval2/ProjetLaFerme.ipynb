```python
def test(df):
    if df['precipitation'] > 5500:
        df['precipitation'] = pluie.loc[(pluie['annee'] == df.annee) & (pluie['saison'] == df.saison) 
        & (pluie['precipitation'] < 5500)].precipitation.mean()
    return df;
```

```python
pluie = pluie.apply(test, axis = 1)
pluie['precipitation'] = pluie['precipitation'].astype('int64')
```
